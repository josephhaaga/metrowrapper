```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
raw=pd.read_json('arrivals_data/arrivals.json')
```

```python
# keys for dataframe columns
raw['Rockville'].dropna()[1].keys()
cleaned_records=pd.DataFrame(columns=raw['Rockville'].dropna()[1].keys())
```

```python
metro_center_arrivals = raw['Metro Center'].dropna()
```

```python
clean_metro_center_arrivals = metro_center_arrivals.reset_index()
clean_metro_center_arrivals = clean_metro_center_arrivals.drop('index',axis=1)
clean_metro_center_arrivals['Metro Center'][1]
```

```python
df = pd.DataFrame(columns=clean_metro_center_arrivals['Metro Center'][1].keys())
```

```python
i=0
for r in clean_metro_center_arrivals['Metro Center']:
    df.loc[i]=r
    i=i+1;
```
